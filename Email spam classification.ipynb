```python
# import the necessary libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
# read the dataset
# Load the email spam dataset
df = pd.read_csv('spam.csv', encoding='latin-1')  # Replace ';' with the actual delimiter.
df = df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
df.head()
```

```python
# Convert labels into binary (ham=0, spam=1)
# df['label'] = df['label'].map({'ham': 0, 'spam': 1})
df = df.replace(['ham','spam'],[0, 1]) 

df.head()
```

```python
# Renaming columns
df.rename(columns={'v1': 'label', 'v2': 'text'}, inplace=True)
```

```python
df.head()
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.3, random_state=42)
```

```python
# Convert text data into numerical format using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
vectorizer
```

```python
# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
# Predict on new test data
new_messages = ['Win cash now!', 'How is your day?', 'Claim your free gift!']
new_features = vectorizer.transform(new_messages)
predictions = model.predict(new_features)
```

```python
# Display results
for message, label in zip(new_messages, predictions):
    print(f"Message: '{message}' -> Predicted Label: '{label}'")
```
